```python
import csv
with open('dow_jones_index.data', newline='') as csvfile:
    datareader = csv.DictReader(csvfile)
    

# Our dictionary, datareader needs help
# new_data = [ {'stock': 'aa', 'close': '$16.42'}, {'stock': 'aa', 'close': '$14.53'}]

    new_data = []

    for row in datareader:
        if row['stock'] not in (d['stock'] for d in new_data):
            new_data.append(row)
        
for stock in new_data:
    print(stock['stock'], stock['close'])

    


hist_data = [
    go.Histogram(
        histfunc = "sum",
        x=[this['stock'] for this in new_data], 
        y=[float(z['close'].strip('$')) for z in new_data]
    )
]
iplot(hist_data)
```
